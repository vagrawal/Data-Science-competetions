```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("train.csv")
train["Semana"] = train["Semana"].astype("category")
train["Agencia_ID"] = train["Agencia_ID"].astype("category")
train["Canal_ID"] = train["Canal_ID"].astype("category")
train["Ruta_SAK"] = train["Ruta_SAK"].astype("category")
train["Cliente_ID"] = train["Cliente_ID"].astype("category")
train["Producto_ID"] = train["Producto_ID"].astype("category")
train.describe(include = 'all')
```

```python
test = pd.read_csv("test.csv")
test["Semana"] = test["Semana"].astype("category")
test["Agencia_ID"] = test["Agencia_ID"].astype("category")
test["Canal_ID"] = test["Canal_ID"].astype("category")
test["Ruta_SAK"] = test["Ruta_SAK"].astype("category")
test["Cliente_ID"] = test["Cliente_ID"].astype("category")
test["Producto_ID"] = test["Producto_ID"].astype("category")
test.describe(include = 'all')
```

```python
train.to_pickle("train.pickle")
test.to_pickle("test.pickle")
```
