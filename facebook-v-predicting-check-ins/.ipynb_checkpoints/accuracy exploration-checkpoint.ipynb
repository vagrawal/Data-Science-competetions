```python
%matplotlib inline
import pandas as pd
import numpy as np
import datetime
import time
import os
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
fig, axs = plt.subplots(3,3, figsize=(50, 30))
axs = axs.ravel()
j = 0

for i, _ in train["place_id"].value_counts()[:1000].sample(9).iteritems():
    df_this = train[train["place_id"] == i]
    
    axs[j].set_title(i)
    axs[j].scatter(df_this["accuracy"], df_this["time"], s = 20, lw = 0, alpha = 0.1);
    j = j + 1
```

```python
fig, axs = plt.subplots(3,3, figsize=(50, 30))
axs = axs.ravel()
j = 0

ranges = ((0, 40), (40, 140), (140, 190), (190, 220), (220, 10000))

for i, _ in train["place_id"].value_counts()[:1000].sample(9).iteritems():
    df_this = train[train["place_id"] == i]
    df_ser = {}
    for l, r in ranges:
        ddd = df_this[(train["accuracy"] >= l) & (train["accuracy"] < r)]
        #print(ddd.shape[0])
        if ddd.shape[0] > 30:
            #print(1)
            df_ser[l] = ddd["y"].var() * 100
    axs[j].set_title(i)
    axs[j].scatter(list(df_ser.keys()), list(df_ser.values()), s = 100)
    j = j + 1
```

```python
fig, axs = plt.subplots(3,3, figsize=(50, 30))
axs = axs.ravel()
j = 0


for i, _ in train["place_id"].value_counts()[:1000].sample(9).iteritems():
    df_this = train[train["place_id"] == i]
    df_ser = {}
    for k in range(0, 200, 10):
        ddd = df_this[(train["accuracy"] >= k) & (train["accuracy"] < k + 10)]
        #print(ddd.shape[0])
        if ddd.shape[0] > 10:
            #print(1)
            df_ser[k] = ddd["y"].mean()
    axs[j].set_title(i)
    axs[j].scatter(list(df_ser.keys()), list(df_ser.values()), s = 100)
    j = j + 1
```

```python
fig, ax = plt.subplots(figsize=(30, 20))
ax.set_xticks(range(0, 250, 10))

xs = np.linspace(1,250,2000)
plt.plot(xs, gaussian_kde(train["accuracy"], bw_method = 0.05)(xs))
```

```python
fig, axs = plt.subplots(3,3, figsize=(50, 30))
axs = axs.ravel()
j = 0

for i, _ in train["place_id"].value_counts()[:1000].sample(9).iteritems():
    df_this = train[train["place_id"] == i]
    #axs[j].hist(df_this["time"], 20, histtype = 'step');
    xs = np.linspace(1,250,2000)
    axs[j].plot(xs, gaussian_kde(df_this["accuracy"], bw_method = 0.05)(xs))
    j = j + 1
```

```python
fig, axs = plt.subplots(3,3, figsize=(50, 30))
axs = axs.ravel()
j = 0

for i, _ in train["place_id"].value_counts()[:1000].sample(1).iteritems():
    df_this = train[train["place_id"] == i]
    #axs[j].hist(df_this["time"], 20, histtype = 'step');
    xs = np.linspace(1,250,2000)
    for j in range(9):
        df_this2 = df_this[(df_this["time"] > j * 100000) & (df_this["time"] < (j + 1) * 100000)]
        axs[j].plot(xs, gaussian_kde(df_this2["accuracy"], bw_method = 0.05)(xs))
```

```python
from pandas.tools.plotting import autocorrelation_plot
fig, axs = plt.subplots(3,3, figsize=(50, 30))
axs = axs.ravel()
j = 0

for i, _ in train["place_id"].value_counts()[:9].iteritems():
    df_this = train[train["place_id"] == i]
    autocorrelation_plot(df_this["time"], ax = axs[j])
    j = j + 1
```
